# Handling the data

```python
#Reading the data from a CSV file
data_file = pd.read_csv("Weather Dataset.csv")

#Handling duplicates
data_file.drop_duplicates(inplace=True)

#Handling missing data
data_file.fillna(method='ffill', inplace=True)

#handling incorrect data
data_file = data_file[data_file["Temperature (C)"] >= -30]
data_file = data_file[data_file["Humidity"] >= 0]


# constant_temp = data_file['Temperature (C)'].value_counts().idxmax()
# constant_periods = data_file[data_file['Temperature (C)'] == constant_temp]

data_file = data_file[data_file['Temperature (C)'] != constant_temp]

non_numeric_columns = data_file.select_dtypes(exclude=['number']).columns
```

# Plotting Temperature over Time

```python
#PLotting temperature over time
plt.figure(figsize=(16, 10))
plt.plot(data_file["Formatted Date"], data_file["Temperature (C)"])
plt.title("Temperature over Time")
plt.ylabel("Temperature")
plt.xlabel("Time")
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=10))
plt.show()
```

# Temperature Histogram

```python
#Temperature Histogram
plt.figure(figsize=(16, 10))
plt.hist(data_file["Temperature (C)"], bins = 10, color = "green")
plt.xlabel("Temperature")
plt.ylabel("Days")
plt.show()
```

# Scatter plot btween Temperature and Humidity

```python
#Scatter plot of Temperature vs Humidity
plt.figure(figsize=(16, 10))
plt.scatter(data_file["Temperature (C)"], data_file["Humidity"], s = 5, c = "red")
plt.title("Temperature vs. Humidity")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()
```

# Correlation Heatmap
The correlation values range from -1 to 1, where -1 indicates a strong negative correlation, 0 indicates no correlation, and 1 indicates a strong positive correlation.

```python
#Correlation heatmap
#print("Non-numeric columns:", non_numeric_columns)

# Step 2: Drop non-numeric columns
data_file_numeric = data_file.select_dtypes(include=['number'])

# Step 3: Compute the correlation matrix
correlation_matrix = data_file_numeric.corr()

# Step 4: Create the heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
